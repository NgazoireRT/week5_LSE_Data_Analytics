```python
# import the requests and Beautiful Soup libraries

import requests
import bs4
from bs4 import BeautifulSoup

# Specify the URL
URL = 'https://en.wikipedia.org/wiki/Main_Page'

# Create a variable
page = requests.get(URL)

# view the HTML
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.text)
```

# Title of the page 

Beautiful Soup can return the title of the website with one code snippet. Usually, there is only one title per website. Therefore, you can use only this simple code snippet:

```python
# Determine the title of the page.
soup.title
```

# Headlines of the page

In general, there is more than one paragraph on a website. Therefore, the find_all() function will give a more accurate output. The code snippet can be written as:

- Define soup.
- Add the find_all() function and specify p within quotation marks.
- Execute the code.

```python
# Find all the main headings on the website.
soup.find_all('h1')
```

# Paragraphs of the page

There is more than one paragraph on a website. Therefore, the find_all() function will give a more accurate output. The code snippet can be written as:

- Define soup.
- Add the find_all() function and specify p within quotation marks.
- Execute the code.

```python
# Find all paragraphs.
soup.find_all('p')
```

# Links on the page 

Looking at the website, it seems there are various links on the page. Let's use a for loop statement to return all the links. The code snippet can be written as:

- Define the for keyword and specify the condition as link in souplfind_all('a):.
- Add a print() function and specify the link and the get() method as link.get('href'). The href is the HTML tag for - links.
- Execute the code.

```python
# Create a for loop to find all the 'a' tags.

for link in soup.find_all('a'):
    print(link.get('href'))
```

# Extracting all the text

You can instruct Python to only return the text without all the HTML tags. The code snippet can be written as:

- Use the print() function and specify soup.get_text method.
- Execute the code.


```python
# Extracting all the text from a page.
print(soup.get_text())
```

# Image alt text
The code snippet for returning the alt text (alternative text and information) of an image can be written as:

- Use the soup.find_all method. and specify images as 'img'.
- Execute the code.

```python
# Return all alt text of the images.
soup.find_all('img')
```
